```python
# https://fr.wikipedia.org/wiki/
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm_notebook as tqdm
import re
import numpy as np
```

```python
tqdm().pandas()
```

```python
def search_for_links(paragraphes_data):
    results = paragraphes_data.find_all(href=re.compile("^/wiki/[a-zA-Z0-9\_\%\(\)\,]*$"), attrs={ 'id': None})
    while len(results)==0:
        paragraphes_data = paragraphes_data.find_next_sibling()
        results = paragraphes_data.find_all(href=re.compile("^/wiki/[a-zA-Z0-9\_\%\(\)\,]*$"), attrs={ 'id': None})
   
    return results
```

```python
def distance_to_philo_article(url='/wiki/Python', rand=False):
    if rand:
        url = requests.get('https://fr.wikipedia.org/wiki/Sp%C3%A9cial:Page_au_hasard').url
        print(url)
    list_visited_link = [url.replace('https://fr.wikipedia.org', '')]
    while list_visited_link[-1] != '/wiki/Philosophie':
        html = requests.get('https://fr.wikipedia.org' + list_visited_link[-1] ).text
        soup = BeautifulSoup(html)

        paragraphes = soup.find('div', class_="mw-parser-output").find(['p','ul'], attrs={'class': None, 'id': None},recursive=False)
        link = search_for_links(paragraphes)

        next_link_url = link[0]['href']
        while next_link_url in list_visited_link or 'API_' in next_link_url :
            link.pop(-len(link))
            if len(link)==0:
                link = search_for_links(paragraphes)
            next_link_url = link[0]['href']
            
        list_visited_link.append(next_link_url)
    return (len(list_visited_link)- 1, list_visited_link)
```

```python
lafayette = distance_to_philo_article('https://fr.wikipedia.org/wiki/Lafayette')
print(lafayette[0])
```
